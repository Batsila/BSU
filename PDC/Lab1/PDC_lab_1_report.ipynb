Результаты экспериментов представить в виде графиков и (или) таблиц. 

```python
import pandas as pd

results = pd.read_csv("test_500.csv")
results1500 = pd.read_csv("test_1500.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt
# plt.rcParams['figure.figsize'] = [10, 8]


def plot_table(results):
    for pos in ['right','top','bottom','left']:
        plt.gca().spines[pos].set_visible(False)
    plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
    plt.tick_params(axis='y', which='both', right=False, left=False, labelleft=False)

    the_table = plt.table(cellText=results.values, colLabels=results.columns, loc='center')
    the_table.auto_set_font_size(False)
    the_table.set_fontsize(24)
    the_table.scale(4, 4)
    plt.show()
```

```python
from IPython.display import display, Markdown

display(Markdown('## Matrix 500 * 500'))
plot_table(results)
display(Markdown('## Matrix 1500 * 1500'))
plot_table(results1500)
```

```python
def plot_graph(df):
    ax = plt.figure()
    x = np.array(df['block_size'])
    y1 = np.array(df['sequential'])
    y2 = np.array(df['first_parallel'])
    y3 = np.array(df['second_parallel'])
    plt.plot(x, y1, 'r', label='sequential', marker='o', linestyle='dashed')
    plt.plot(x, y2, 'g', label='first_parallel', marker='o', linestyle='dashed')
    plt.plot(x, y3, 'b', label='second_parallel', marker='o', linestyle='dashed')
    ax.legend()
    plt.show()
```

```python
display(Markdown('## Matrix 500 * 500'))
plot_graph(results)
```

```python
display(Markdown('## Matrix 1500 * 1500'))
plot_graph(results1500)
```

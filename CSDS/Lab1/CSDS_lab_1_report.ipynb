```python
import pandas as pd

results_text = pd.read_csv('text.csv', ';')
results_key = pd.read_csv('key.csv', ';')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_graph(data_frame, x_name, y_name, plot_name):
    x = data_frame[x_name]
    y = [float(num.replace(',', '.')) for num in data_frame[y_name]]
    plt.plot(x, y, 'g', marker='o', linestyle='dashed')
    plt.show()
```

```python
results_key_250 = results_key[0 : 20]
plot_graph(results_key_250, 'KeyLength', 'Probability', 'Key 250')
```

```python
results_key_500 = results_key[20 : 40]
plot_graph(results_key_500, 'KeyLength', 'Probability', 'Key 500')
```

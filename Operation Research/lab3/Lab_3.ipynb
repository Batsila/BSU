```python
import numpy
infinity = 42000

def print_vector(vector):
    str_line = ""
    for e in vector:
        if (e is infinity):
            str_line += 'inf '
        else:
            str_line += str(e) + ' '
    print(str_line)

def print_matrix(matrix):
    for line in matrix:
        str_line = ""
        for e in line:
            if (e is infinity):
                str_line += 'inf '
            else:
                str_line += str(e) + ' '
        print(str_line)
```

Алгоритм Флоида

```python
d = [
    [0, 11, 2, 8, 11, 11],
    [11, 0, 5, infinity, infinity, 1],
    [infinity, 5, 0, infinity, 2, 1],
    [2, infinity, infinity, 0, 2, infinity],
    [infinity, 9, infinity, 2, 0, 7],
    [infinity, 1, infinity, infinity, 7, 0]
]

t = [
    [0, 2, 3, 4, 5, 6],
    [1, 0, 3, 4, 5, 6],
    [1, 2, 0, 4, 5, 6],
    [1, 2, 3, 0, 5, 6],
    [1, 2, 3, 4, 0, 6],
    [1, 2, 3, 4, 5, 0]
]

print('D0')
print_matrix(d)
print('T0')
print_matrix(t)
print()

for k in range(6):
    for i in range(6):
        for j in range(6):
            if i is not j and i is not k and j is not k:
                if d[i][k] + d[k][j] < d[i][j]:
                    d[i][j] = d[i][k] + d[k][j]
                    t[i][j] = k + 1
    print('D{}'.format(k + 1))
    print_matrix(d)
    print('T{}'.format(k + 1))
    print_matrix(t)
    print()
```

Алгоритм Форда-Беллмана

```python
d = numpy.array([
    [0, -1, infinity, 3, infinity, infinity, infinity, infinity, infinity],
    [-1, 0, 1, infinity, 2, infinity, infinity, infinity, infinity],
    [infinity, 1, 0, infinity, -3, 6, infinity, infinity, infinity],
    [3, infinity, infinity, 0, -3, infinity, 5, infinity, infinity],
    [infinity, 2, -3, -3, 0, infinity, -2, -1, infinity],
    [infinity, infinity, 4, infinity, infinity, 0, infinity, infinity, -3],
    [infinity, infinity, infinity, 5, -2, infinity, 0, 3, infinity],
    [infinity, infinity, infinity, infinity, -1, infinity, 3, 0, 4],
    [infinity, infinity, infinity, infinity, infinity, -3, infinity, 4, 0]
])

dist = []
dist.append(0)
while len(dist) is not 9:
    dist.append(infinity)

print_vector(dist)
for k in range(9):
    for i in range(9):
        for j in range(9):
            if dist[i] > dist[j] + d[i][j]:
                dist[i] = dist[j] + d[i][j]
    print_vector(dist) 
          
```
